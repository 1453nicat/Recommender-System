```python
from pandas import *

links = read_csv("/content/links.csv")
movies = read_csv("/content/movies.csv")
ratings = read_csv("/content/ratings.csv")
tags = read_csv("/content/tags.csv")

print(links)
print(movies)
print(ratings)
print(tags)
```

```python
#merge movies and ratings

data = merge(movies, ratings, on = 'movieId')
data.head()
```

```python
data = data[["userId", "title", "rating"]]
pivot = data.pivot_table(index = 'title', columns = 'userId', values ='rating', fill_value = 0)

(pivot)
```

```python
#correlation
from sklearn.metrics import pairwise_distances

matrix = pairwise_distances(pivot, metric = "correlation")
print(matrix)
```

```python
#similar films as "Godfather, The (1972)"
indeces = list(pivot.index)
film_index = indeces.index("Godfather, The (1972)")

sim_films = (matrix[film_index].argsort()[1 : 5])
for i in sim_films:
  print(indeces[i])
```
